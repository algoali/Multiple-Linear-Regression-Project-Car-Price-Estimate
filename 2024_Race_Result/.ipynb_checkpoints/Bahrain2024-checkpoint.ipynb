### Bahrain 2024

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
import csv
```

```python
url = "https://www.formula1.com/en/results.html/2024/races/1229/bahrain/race-result.html"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
table = soup.find_all('table', class_='resultsarchive-table')
```

```python
df =pd.read_html(str(table))[0]
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
df.to_csv('2024_CSV/Bahrain2024_result.csv', index=False)
```

```python
# Practice 1
url = "https://www.formula1.com/en/results.html/2024/races/1229/bahrain/practice-1.html"
response = requests.get(url)
response.status_code
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find_all('table', class_='resultsarchive-table')
df =pd.read_html(str(table))[0]
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.to_csv('2024_CSV/Bahrain2024_p1.csv', index=False)
df
```
